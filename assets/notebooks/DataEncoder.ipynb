```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno
from skimpy import skim
import warnings
import os
import pickle
import plotly.express as px
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

## Load the data

```python
#  Training
skim(training_df)
```

## Encoder

```python
# Identificar las columnas categórica y numericas
object_cols = list(training_df.select_dtypes(include='object').columns)
number_cols = list(training_df.select_dtypes(exclude='object').columns)
```

```python
# Obtener X y Y
X = training_df.drop(['DiagPeriodL90D','patient_id'],axis = 1)
ids = training_df['patient_id']
Y = training_df['DiagPeriodL90D']
```

```python
# Encoder
from category_encoders import CatBoostEncoder
number_cols.remove('DiagPeriodL90D')
number_cols.remove('patient_id')
encoder = CatBoostEncoder(cols=object_cols)
X_objects_col = encoder.fit_transform(X[object_cols], Y)
X = pd.concat([X_objects_col.reset_index(drop=True), X[number_cols].reset_index(drop=True)], axis=1)
```

```python
# Guardar Modelos
import os
import pickle
pickle.dump(encoder, open(os.path.join('/content/drive/MyDrive/WIDS/MODELS/Encoder.sav'), 'wb'))
print('Se ha guardado exitosamente')
```

```python
df_encoder = pd.concat([ids,X,Y],axis=1)
df_encoder
```
